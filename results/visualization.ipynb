```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv('results.csv')

# Display the first few rows of the data
print(data.columns)

```

```python
# Visualise the data based on Running Time
# For each batch number 

plt.figure(figsize=(10, 6))
plt.plot(data[(data['BatchSize'] == 64) & (data['TestRun'] == 2)]['Threads'], np.log(data[(data['BatchSize'] == 64) & (data['TestRun'] == 2)]['Time(ms)']), marker='o', linestyle='-', color='b', label='Batch 64')
plt.plot(data[data['BatchSize'] == 128]['Threads'], np.log(data[data['BatchSize'] == 128]['Time(ms)']), marker='o', linestyle='-', color='r', label='Batch 128')
plt.plot(data[data['BatchSize'] == 512]['Threads'], np.log(data[data['BatchSize'] == 512]['Time(ms)']), marker='o', linestyle='-', color='y', label='Batch 512')
plt.xlabel('Number of Threads')
plt.ylabel('LogRunning Time (seconds)')
plt.title('Running Time vs Number of Threads')
plt.grid(True)
plt.legend()
plt.show()

# Visualise the data based on SpeedUp
plt.figure(figsize=(10, 6))
plt.plot(data[(data['BatchSize'] == 64) & (data['TestRun'] == 2)]['Threads'], data[(data['BatchSize'] == 64) & (data['TestRun'] == 2)]['SpeedUp'], marker='o', linestyle='-', color='b', label='Batch 64')
plt.plot(data[data['BatchSize'] == 128]['Threads'], data[data['BatchSize'] == 128]['SpeedUp'], marker='o', linestyle='-', color='r', label='Batch 128')
plt.plot(data[data['BatchSize'] == 512]['Threads'], data[data['BatchSize'] == 512]['SpeedUp'], marker='o', linestyle='-', color='y', label='Batch 512')
max_threads = data['Threads'].max()
plt.plot([1, max_threads], [1, max_threads], 'b--', label='Ideal Linear Speedup')

plt.xlabel('Number of Threads')
plt.grid(True)
plt.legend()
plt.ylabel('SpeedUp')
plt.title('SpeedUp vs Number of Threads')
```
